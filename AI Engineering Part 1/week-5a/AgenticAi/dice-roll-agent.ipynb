```python
import random
from openai import OpenAI

client = OpenAI()

def roll_dice():
    return random.randint(1,6)
```

```python
target = 6
max_iterations = 20
```

```python
from matplotlib.pyplot import hist


history = ""
iteration = 0

while iteration < max_iterations:
    iteration += 1

    instruction = f""" 
    You are playing a dice game. You need to roll a {target} on a dice.

    You have made the following so far:
    {history if history else "No rolls yet."}

    Based on what you have so far, decide your next action: 

    Option A: if you have achieved the target, response with exactly: DONE: message 
              describing how you achieved the target. 

    Option B: if you have not achieved the target, response with exactly: ROLL A DICE
"""

    response = client.chat.completions.create(
        model = "gpt-4.1-mini",
        messages = [{"role":"user", "content": instruction}]
    )

    result = response.choices[0].message.content

    if "DONE" in result:
        print(result)
        break

    if "ROLL A DICE" in result:
        roll = roll_dice()
        print(f"Dice rolled: {roll}\n")

        history += f"Roll {iteration}: {roll}\n"


print(f"Game ended after {iteration-1} iterations")
```
