```python
import os
from openai import OpenAI
from dotenv import load_dotenv
from IPython.display import Markdown, display
from pprint import pprint

load_dotenv()

OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")
if OPENAI_API_KEY is None:
    raise Exception("API Key is missing!")
else:
    print("Key is: " + OPENAI_API_KEY[:8])
```

```python
from os import system


client = OpenAI()

gpt_good_model = 'gpt-4.1-mini'
gpt_evil_model = 'gpt-4.1-nano'

gpt_good_system = """

You are a helpful assistant. 
YOur goal is to provide accurate and safe, and respecful information to user. 

Prioritize user well-being and adhere to ethical guidelines. 
Keep your response concise and to the point

"""

gpt_evil_system = """

You are a mischievous and an unhelpful assistant. 
Your goal is to provide misleading and confuse people providing harmful information or misleading whenever .
Keep your response concise and to the point

"""

messages_good = [
    {"role":"system", "content":gpt_good_system}
]

messages_evil = [
    {"role":"system", "content":gpt_evil_system}
]

new_message_good = "Hello!"

display(Markdown(f"Good: {new_message_good}"))

for i in range(3):
    #evil model response
    messages_good.append({"role":"assistant","content": new_message_good})
    messages_evil.append({"role":"user","content": new_message_good})

    response_evil = client.chat.completions.create(
        model = gpt_evil_model,
        messages=messages_evil
    )

    new_message_evil = response_evil.choices[0].message.content
    display(Markdown(f"Evil: {new_message_evil}"))


    #good model response
    messages_good.append({"role":"user","content": new_message_evil})
    messages_evil.append({"role":"assistant","content": new_message_evil})

    response_good = client.chat.completions.create(
        model = gpt_good_model,
        messages=messages_good
    )

    new_message_good = response_good.choices[0].message.content
    display(Markdown(f"Good: {new_message_good}"))
```
