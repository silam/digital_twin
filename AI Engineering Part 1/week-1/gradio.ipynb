### Step 1: Import libraries

```python
import os
from openai import OpenAI
from dotenv import load_dotenv

load_dotenv()

OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")
if OPENAI_API_KEY is None:
    raise Exception("API Key is missing!")
else:
    print("Key is: " + OPENAI_API_KEY[:8])
```

### Step 2: Basic User Interface

```python
import gradio as gr
```

```python
import datetime
from pprint import pprint

from click import DateTime
from datetime import datetime
from openai import OpenAI

def response_ai(message, history):
    client = OpenAI(api_key=OPENAI_API_KEY)
    messages = [{"role":"system", "content": "You are an extremely unhelpful assistant"}] + history + [{"role":"user", "content": message}]
    
    response = client.chat.completions.create(
        model = "gpt-4.1-mini",
        messages= messages
    )
    answer = response.choices[0].message.content

    return answer
```

```python
from gradio import Chatbot

gr.ChatInterface(fn=response_ai, autofocus=True, autoscroll=True, submit_btn=True, examples=["hello", "hola", "merhaba"], title="Echo Bot").launch(inbrowser=True) #, share=True)
```
