# Step 1: Import libraries

```python
import os
from dotenv import load_dotenv
from openai import OpenAI
from IPython.display import display, Markdown

load_dotenv()

OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")

if OPENAI_API_KEY is None:
    raise Exception("API Key is missing!")

```

### Step 2: Searching the web

```python
client = OpenAI()

completion = client.chat.completions.create(
    model="gpt-4o-mini-search-preview",
    messages=[
        {"role": "user","content":"Tell me the war in Middle East today? List the sources. Don't use simple news aggregators"}
    ],
    max_tokens=500
)
```

```python
search_results = completion.choices[0].message.content

display(Markdown(search_results))
```

```python
system_message = """ 
You are amazing helpful assistant who creates amazing linkedin posts about AI based
on text you are provided with. Make the post engaging and interesting.
Also make concise and add a thought questions at the end to encourage comments.


"""

prompt = f""" 
The following contains the news today
Now please write my linkedin post based on this information.

{search_results}

"""

completion = client.chat.completions.create(
    model="gpt-4o-mini",
    messages = [
        {"role": "system", "content": system_message},
        {"role": "user", "content": prompt}
    ],
    max_tokens=500,
    temperature=0.5
)

chat_respone = completion.choices[0].message.content

display(Markdown(chat_respone))
```
